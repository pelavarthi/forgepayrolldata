```python
query1 = """
SELECT employees.employee_id,employees.full_name,employees.department_id,employees.manager_id,employees.job_title,employees.state,departments.department_name,departments.location,departments.budget,salaries.base_salary,salaries.bonus,salaries.deductions,taxes.tax_rate,taxes.num_exemptions,taxes.estimated_withholding_amt
FROM employees
INNER JOIN departments
    ON employees.department_id=departments.department_id
INNER JOIN salaries
    ON employees.employee_id=salaries.employee_id
INNER JOIN taxes
    ON employees.employee_id=taxes.employee_id;
"""
```

```python
result1 = pd.read_sql(query1, con=connection)
result1
```

```python
query3 = """
SELECT full_name, job_title, base_salary+bonus-deductions-estimated_withholding_amt AS net_income
FROM payroll
ORDER BY net_income DESC
"""
```

```python
query4 = """
SELECT full_name, job_title, round((base_salary+bonus-deductions-estimated_withholding_amt)/(base_salary+bonus)*100) AS take_home_perc
FROM payroll
ORDER BY base_salary DESC
"""
```

```python
query5 = """
SELECT department_name, round(AVG(base_salary+bonus)) AS average_gross_pay
FROM payroll
GROUP BY department_name
"""
```

```python
result5 = pd.read_sql(query5, con=connection)
result5
```

```python
query6="""
SELECT state, 
       AVG(base_salary) AS median_base_salary
FROM (
    SELECT state, base_salary,
           ROW_NUMBER() OVER (PARTITION BY state ORDER BY base_salary) AS rn,
           COUNT(*) OVER (PARTITION BY state) AS cnt
    FROM payroll
) AS ranked
WHERE rn IN (FLOOR((cnt + 1)/2), FLOOR((cnt + 2)/2))
GROUP BY state;
"""
```

```python
result6 = pd.read_sql(query6, con=connection)
result6
```

```python
query7="""
SELECT job_title, 
       AVG(base_salary) AS median_base_salary
FROM (
    SELECT job_title, base_salary,
           ROW_NUMBER() OVER (PARTITION BY job_title ORDER BY base_salary) AS rn,
           COUNT(*) OVER (PARTITION BY job_title) AS cnt
    FROM payroll
) AS ranked
WHERE rn IN (FLOOR((cnt + 1)/2), FLOOR((cnt + 2)/2))
GROUP BY job_title
ORDER by median_base_salary DESC;
"""
```

```python
result7 = pd.read_sql(query7, con=connection)
result7
```
